```python
num = df.select_dtypes(include="number")
summary = num.agg(["mean", "median"]).T
summary["mode"] = num.mode(dropna=True).iloc[0]
summary
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
outlier_counts = {}
for col in num.columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    low = q1 - 1.5 * iqr
    high = q3 + 1.5 * iqr
    outlier_counts[col] = ((df[col] < low) | (df[col] > high)).sum()

pd.Series(outlier_counts).sort_values(ascending=False)
```

Using the outlier IQR test, doesn't seem to display any outliers in the data.
