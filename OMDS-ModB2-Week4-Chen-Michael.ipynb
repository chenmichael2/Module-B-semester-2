```python
num_data = df.select_dtypes(include="number")
num_data
```

```python
num_data.hist(figsize=(12, 10), bins=20)
```

```python
mean = df[["MartianRide", "TeacupRide", "RiverRide"]].mean().to_list()
median = df[["MartianRide", "TeacupRide", "RiverRide"]].median().to_list()
mode = df[["MartianRide", "TeacupRide", "RiverRide"]].mode().iloc[0]
pd.DataFrame([mean, median, mode], index=["Mean", "Median", "Mode"], columns=["Martian", "Teacup", "River"])
```

```python
mean = num_data.mean().to_list()
median = num_data.median().to_list()
mode = num_data.mode().iloc[0]
std = num_data.std().to_list()
var = num_data.var().to_list()
pd.DataFrame([mean, median, mode, std, var], index=["Mean", "Median", "Mode", "Std", "Variance"], columns=num_data.columns)
```
